## Políticos check, antecedentes dos parlamentares



###### Análise do Congresso brasileiro com foco em processos judiciais que os parlamentares respondem no Supremo Tribunal Federal e em tribunais estaduais e eleitorais. A fonte primária de dados foi o site http://www.politicos.org.br/processos que acompanha esse registros. Foi feita uma raspagem com planilhas do Google, limpeza e estruturação dos dados das ocorrências por agrupamento dos principais crimes contra a <i>Res publica</i>, a coisa pública que é de todos que produzem as riquezas administradas por entes governamentais. A maior parte do links foi revisitada para checagem, em alguns tribunais estaduais a consulta é dificultada por exigir nome de advogados ou número de processo, não permitindo apenas a busca pelo nome das partes. Esta análise foi concebida durante o MOOC "Como cobrir eleições sem errar: dados e pesquisas para entender o eleitor", oferecido pelo Centro Knight para o Jornalismo nas Américas e a Associação Nacional dos Jornais (ANJ), em parceria com o Google News Initiative.
