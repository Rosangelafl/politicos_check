### Análise dos crimes ambientais no Congresso. Além dos parlamentares que respondem ao crime de Dano Ambiental, os que possuem áreas embargadas pelo Ibama e os que receberam nas eleições de 2014 doações de empresas flagradas por trabalho análogo a escravidão. As fontes de dados são os sites http://www.politicos.org.br/processos e as planilhas disponibilizadas pelo https://ruralometro.reporterbrasil.org.br.


```python
import pandas as pd
```

```python
ambiental = pd.read_csv('politicos_check_trabalho.csv')
```

```python
ambiental.head()
```

```python
ambiental.info()
```

```python
ambiental[ambiental.ocorrencia == 'Dano Ambiental']
```

```python
crimes = ambiental[ambiental.ocorrencia == 'Dano Ambiental']
```

#### 20 deputados e senadores respondem por dano ambiental. MDB, PP e SD são os partidos com mais infratores com 3 parlamentares cada. Seguido por PSD, PSDB e PT com 2. O estado de São Paulo é o que tem mais parlamentares que cometeram danos ambientais com  4 integrantes, seguido por Minas Gerais, Pernanbuco e Ceára com 3 integrantes cada.
